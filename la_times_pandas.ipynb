# Import libraries and data
Use pandas "read_csv" method to import spreadsheets and store them as variables.

```python
import pandas as pd
```

```python
%matplotlib inline
```

```python
commits = pd.read_csv("http://www.firstpythonnotebook.org/_static/committees.csv")
```

```python
contribs = pd.read_csv("http://www.firstpythonnotebook.org/_static/contributions.csv")
```

# Value counts and reset index
Target specific column with ".column_name". Use pandas method, "value_counts", to tally counts for each value in the column, use reset index to transform from series to DataFrame.

```python
commits.prop_name.value_counts().reset_index()
```

# Filtering a dataframe by a field value
Result will be a new data frame which you should store in a variable. Format below, dataframe column equal to desired field value.

```python
weedcommits= commits[commits.prop_name == 'PROPOSITION 064- MARIJUANA LEGALIZATION. INITIATIVE STATUTE.']
```

```python
weedcommits.head()
```

```python
weedcommits.info()
```

```python
contribs.head()
```

# Using pandas "merge" method to filter
The default merge method in pandas is an inner join, so only the fields with matching values for the specified column will be retained in the new data frame. Store merged dataframe in a new variable. Duplicated columns will have an index to specify their origin. Inputs of the merge method are the dataframes to be joined and the field to join them on.

```python
merged = pd.merge(weedcommits, contribs, on="calaccess_committee_id")
```

```python
merged.head()
```

```python
merged.info()
```

# Analyze Data
Target campaign contributions amount column and sum up. Separate support and oppose contributors by filtering.

```python
merged.amount.sum()
```

```python
merged.committee_position.value_counts()
```

```python
support = merged[merged.committee_position == 'SUPPORT']
```

```python
oppose = merged[merged.committee_position == 'OPPOSE']
```

```python
support.amount.sum()
```

```python
oppose.amount.sum()
```

```python
support.amount.sum() / oppose.amount.sum()
```

# Using the sort method
Identify largest donors using pandas sort method. Inputs are column to sort, default is ascending.

```python
merged.sort_values('amount', ascending=False).head()
```

```python
oppose.sort_values('amount', ascending=False).head(20)
```

# Use groupby method and chaining for efficiency
The groupby method in pandas is very handy for analyzing categorical data. The input is the column you wish to sort by, then chain on the analysis and display methods. For multiple columns, add as a series with flat brackets, and don't forget quotes!

```python
merged.groupby("committee_name_x").amount.sum().reset_index().sort_values("amount", ascending=False)
```

```python
merged.groupby(["contributor_lastname", "contributor_firstname", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False)
```

```python
top_supporters = support.groupby(["contributor_lastname", "contributor_firstname"]).amount.sum().reset_index().sort_values("amount", ascending=False).head(10)
```

# Create new column
Create full name column by concatenating two other columns with a space between

```python
top_supporters['contributor_fullname'] = top_supporters.contributor_firstname + " " + top_supporters.contributor_lastname
```

# Create chart with matplotlib "chart" method and change yticklabels

```python
chart = top_supporters.head(5).amount.plot.barh()
chart.set_yticklabels(top_supporters.contributor_fullname)
```

# Export DataFrame to .csv

```python
top_supporters.to_csv("top_supporters.csv")
```
